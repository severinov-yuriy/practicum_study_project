Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

__Описание данных:__

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)

- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — площадь квартиры в квадратных метрах (м²)

### Замена пропусков

Для заполнения пропущенных значений воспользуемся методом .fillna().

```python
real_estate['is_apartment'] = real_estate['is_apartment'].fillna(False)
```

```python
for column in ('balcony', 'ponds_around3000', 'parks_around3000'):
    real_estate[column] = real_estate[column].fillna(0)
```

Все пропуски были заменены.

### Обработка пропусков

Заполним пропуски в высоте потолков средним значением по населенному пункту. Пропуски, которые могли остаться в силу того, что в определенном населенном пункте нет объектов с информацией о высоте потолков заменим средним значением высоты потолка по всей выборке.

```python
ceiling_height_by_town = real_estate.groupby(by='locality_name')['ceiling_height'].describe()['mean'].to_dict()
real_estate['ceiling_height'] = real_estate['ceiling_height'].fillna(real_estate['locality_name'].map(ceiling_height_by_town))
real_estate['ceiling_height'] = real_estate['ceiling_height'].fillna(real_estate['ceiling_height'].mean())
```

Заполним пропуски количества этажей средним значением по населенному пункту. Пропуски, которые могли остаться в силу того, что в определенном населенном пункте нет объектов с информацией о высоте потолков заменим средним значением высоты потолка по всей выборке.

```python
floors_total_by_town = real_estate.groupby(by='locality_name')['floors_total'].describe()['mean'].round().to_dict()
real_estate['floors_total'] = real_estate['floors_total'].fillna(real_estate['locality_name'].map(floors_total_by_town))
real_estate['floors_total'] = real_estate['floors_total'].fillna(real_estate['floors_total'].mean())
```

Проверим, не превысило ли значение этажа, максимальную этажность дома после заполнения.

```python
len((real_estate[real_estate['floors_total'] < real_estate['floor']]))
```

Для 29 объектов заполненный по среднему максимальный этаж стал меньше указанного этажа в доме. Так как таких значений немного, возьмем указанный в датасете этаж за максимальный, что практически не исказит итоговый результат.

```python
real_estate.loc[real_estate['floors_total'] < real_estate['floor'], 'floors_total'] = real_estate['floor']
```

Проверим, что максимальный этаж не меньше указанного еще раз.

```python
len((real_estate[real_estate['floors_total'] < real_estate['floor']]))
```

Все в порядке, можно переходить к жилой пллощади.

Заполним пропуски жилой площади значением средней доли жилой площади от общей площади.

```python
mean_living_area = (real_estate['living_area'] / real_estate['total_area']).mean()
real_estate['living_area'] = real_estate['living_area'].fillna(mean_living_area)
```

Заполним пропуски площади кухни общим средним значением.

```python
real_estate['kitchen_area'] = real_estate['kitchen_area'].fillna(real_estate['kitchen_area'].mean())
```

Проверим, не превысила  ли суммарная жилая площадь и площадь кухни общую площадь квартиры.

```python
len(real_estate[(real_estate['living_area'] + real_estate['kitchen_area']) > real_estate['total_area']])
```

Для 366 объектов общая площадь квартиры получилась меньше суммарной жилой и пллощади кухни. Для них возьмем жилую площадь как разницу общей площади и площади кухни. Будем считать, что в таких квартирах нет коридора или он учтен в жилой площади и сразу проверим, сработала ли правка.

```python
real_estate.loc[
    (real_estate['living_area'] + real_estate['kitchen_area']) > real_estate['total_area'],
    ['living_area']] = real_estate['total_area'] - real_estate['kitchen_area']
real_estate[(real_estate['living_area'] + real_estate['kitchen_area']) > real_estate['total_area']]
```

Отобразился один объект, для которого площадь кухни и жилая не превышают общую. Вероятно дело в сотых долях, однако это недостаточно значительная ошибка для дальнейшей проработки на итоговый результат она не повлияет, поэтому имеет смысл оставить так как есть.

Все пропуски были обработаны.

### Изменение типа данных

Изменим тип данных в столбце `first_day_exposition`.

```python
real_estate['first_day_exposition'] = pd.to_datetime(real_estate['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Изменим тип данных с вещественного на целочисленный.

```python
columns_list = ('last_price', 'floors_total', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000',
                'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition')
for column in columns_list:
    real_estate[column] = pd.to_numeric(real_estate[column], downcast='integer', errors='coerce')
```

Проверим, во всех ли столбцах тип данных был изменен.

```python
real_estate.info(memory_usage='deep')
```

Тип данных был изменен в части столбцов, однако в столбцах `airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest` и `days_exposition` данные остались в формате float64. Это связано с тем, что в этих столбцах остались пропуски в формате NaN, которые имеют тип данных float64 и не конвертируются в тип int.

### Вывод

Пропуски в данных были заполнены и тип данных был изменен на более подходящий. Это позволило также снизить размер датасета почти на 40%.

## Расчёты и добавление результатов в таблицу

### Постановка задачи

__Для дальнейшего анализа необходимо провести расчеты и добавить в таблицу следующие данные:__

- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

### Расчет

Рассчитаем цену за квадратный метр как отношение цены квартиры к ее площади.

```python
real_estate['price_per_meter'] = real_estate['last_price'] / real_estate['total_area']
```

Из столбца `first_day_exposition` получим данные о дне недели, месяце и годе выставления объекта на продажу.

```python
real_estate['weekday'] = real_estate['first_day_exposition'].dt.dayofweek
real_estate['month'] = real_estate['first_day_exposition'].dt.month
real_estate['year'] = real_estate['first_day_exposition'].dt.year
```

Для определения категории этажа напишем функцию, возвращающую тип этажа на основе значений этажа и этажности дома. 0 будет означать первый этаж, 1 - другой, 2 - последний этаж.

```python
def floor_type(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 0
    elif floor == floors_total:
        return 2
    return 1
real_estate['floor_type'] = real_estate.apply(floor_type, axis=1)
```

Рассчитаем соотношение жилой зоны и зоны кухни к общей площади  и добавим в датасет.

```python
real_estate['living_area_per_total'] = real_estate['living_area'] / real_estate['total_area']
real_estate['kitchen_area_per_total'] = real_estate['kitchen_area'] / real_estate['total_area']
```

Далее выясним, какая область в Санкт-Петербурге считается центром города. Для этого построим график роста средней цены в зависимости от отдаления от центра. Сперва выделим квартиры, которые находятся в Санкт-Петербурге и отбросим квартиры, по которым отсутствует значения расстояния до центра. Также переведем столбец с расстоянием до центра в километры в целых значениях, округлив математичски.

```python
(
    spb_real_estate.groupby(by='km_to_center').mean()['last_price'].plot(grid=True, drawstyle="steps")
    .set(title='Mean price per kilometer',
         xlabel="distance to center, km",
         ylabel="flat price, 10mln rub",
         xticks=(range(0, 30, 8)))
)

plt.show()
```

На графике четко видно, что на расстоянии 8 километров от центра цена резко падает с 9 млн. рублей до 6 млн. рублей. Можно считать, что центром Санкт-Петербурга считается зона, находящаяся в пределах 8 километров от центра.

Для сравнения объектов недвижимости в целом по выборке, в Петербурге и центре Петербурга, сформируем сводную таблицу статистических показателей.

```python
columns_median = {'total_area': 'Медиана площади',
                         'last_price': "Медиана цены",
                         'price_per_meter': "Медиана цены за квадратный метр"}
columns_mean = {'rooms': "Среднее число комнат",
                       'ceiling_height': "Средняя высота потолка",
                       'floors_total': "Средняя высота дома",
                       'living_area': "Средняя жилая площадь",
                       'floor': "Средний этаж в объявлении",
                       'kitchen_area': "Средняя площадь кухни",
                       'parks_nearest': "Средняя удаленность от парков",
                       'ponds_nearest': "Средняя удаленность от водоемов",
                       'days_exposition': "Среднее время продажи"}
whole_data_mean_flat = (
    real_estate.describe().loc[['50%'], columns_median.keys()]
    .rename({"50%": "mean"})
    .join(real_estate.describe().loc[['mean'], columns_mean.keys()])
    .rename({"mean": "В целом по выборке"})
    .T
    .rename(columns_median)
    .rename(columns_mean)
)
spb_data_mean_flat = (
    spb_real_estate.describe().loc[['50%'], columns_median.keys()]
    .rename({"50%": "mean"})
    .join(spb_real_estate.describe().loc[['mean'], columns_mean.keys()])
    .rename({"mean": "По СПб"})
    .T
    .rename(columns_median)
    .rename(columns_mean)
)
center_data_mean_flat = (
    center_real_estate.describe().loc[['50%'], columns_median.keys()]
    .rename({"50%": "mean"})
    .join(center_real_estate.describe().loc[['mean'], columns_mean.keys()])
    .rename({"mean": "По центру СПб"})
    .T
    .rename(columns_median)
    .rename(columns_mean)
)
flats_comparison = whole_data_mean_flat.join(spb_data_mean_flat).join(center_data_mean_flat)
```

### Выводы

Основные параметры, предопределяющие цену квартиры это расстояние до центра, город размещения объекта и ее площадь. Тип этажа влияет на цену квартиры, но незначительно. Прочие параметры, такие как  количество парков, водоемов, свободная планировка и высотность здания также влияют на цену продажи квартиры.
